```python
import pandas as pd
import matplotlib as plt
import regex as re
```

```python
videogame_df = pd.read_csv("Video_Games_Sales_as_at_22_Dec_2016.csv")
videogame_df
```

What is represented on each line?
Each line is a game version of specific franchises depending on the platform it was sold on. The sales in the main video game markets and the critic scores are listed, as well as the developpers and publishers, genre and rating for usage

```python
# Jeu vidÃ©o le plus vendu
print(videogame_df['Global_Sales'].max())
videogame_df.loc[videogame_df['Global_Sales']==videogame_df['Global_Sales'].max()]['Name']


```

```python
# Platforme ayant vendu le plus de jeux
platform_df = videogame_df.groupby("Platform")['Global_Sales'].sum().sort_values(ascending=False)
platform_df
```

```python
# Rating Adults only
print(videogame_df.loc[videogame_df['Rating']=='AO'])
print(videogame_df.loc[videogame_df['Name']== 'Grand Theft Auto V'].groupby(videogame_df['Name']).value_counts())
```

```python
# Most sold game any platform
most_sold_games = videogame_df.groupby("Name")['Global_Sales'].sum().sort_values(ascending=False)
most_sold_games
```

```python
# Jeu PokÃ©mon le plus vendu
pokemon_games = videogame_df[videogame_df['Name'].str.contains('pokemon', case=False, na=False)]
pokemon_games.sort_values('Global_Sales',ascending=False)
```
