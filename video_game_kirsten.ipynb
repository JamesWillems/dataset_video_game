```python
import pandas as pd
import matplotlib as plt
```

```python
videogame_df = pd.read_csv("Video_Games_Sales_as_at_22_Dec_2016.csv")
videogame_df
```

```python
videogame_df["Platform"].value_counts()
```

The data wad found from a web scrap form metacritic

In the different columns we have:
- the name of a game
- the platform it's playable
- the year it was released
- the genre of the game
- franchise of the game
- sales in europe, north america, japan and other countries and in global
- the number of critics and the score
- the users' score and the number of users that gave a score
- the developer of the game
- the rating of the game


The dataset represents a sample of games until 2016

```python
#check if a video game in the dataset is playable on different platforms
videogame_df.loc[videogame_df.duplicated('Name')]
```

```python
#check which game is adult only
videogame_df.loc[videogame_df['Rating'] == 'AO']
```

```python
#check which games are rated mature
videogame_df.loc[videogame_df['Rating'] == 'M']
```

```python
#check which Publisher is the best
videogame_df['Publisher'].value_counts().idxmax()
```

```python
#check how many games are there in the dataset
videogame_df['Name'].nunique()

#check how many games a publisher has without counting duplicates
no_duplicates= videogame_df.drop_duplicates(subset=['Name', 'Publisher'])
no_duplicates['Publisher'].value_counts()
```

```python
#check the best selling game from Namco Bandai Games
videogame_df.loc[videogame_df['Publisher'] == 'Namco Bandai Games'].sort_values('Global_Sales', ascending=False)
```

```python
#check the info of Namco Museum
videogame_df.loc[videogame_df['Name'] == 'Namco Museum']
```

```python
#check the which genre is the most popular
print(videogame_df['Genre'].value_counts())

#check the best selling action game
videogame_df.loc[videogame_df['Genre'] == 'Action'].sort_values('Global_Sales', ascending=False)
```

```python
#check the name of the best selling game by platform
videogame_df.loc[videogame_df.groupby('Platform')['Global_Sales'].idxmax().sort_values(ascending=True)][['Name', 'Platform', 'Global_Sales']]

```

```python
#check the best selling game by region
videogame_df.loc[videogame_df.groupby('EU_Sales')['Global_Sales'].idxmax().sort_values(ascending=True)][['Name', 'Platform', 'Global_Sales']]
```

```python
#check the best selling game by region
videogame_df.loc[videogame_df.groupby('NA_Sales')['Global_Sales'].idxmax().sort_values(ascending=True)][['Name', 'Platform', 'Global_Sales']]
```
